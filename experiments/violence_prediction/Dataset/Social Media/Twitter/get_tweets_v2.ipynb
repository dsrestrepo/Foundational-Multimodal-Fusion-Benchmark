```python
from dotenv import load_dotenv
import os

load_dotenv()

# Set Scraper API key
scraper_api_key = os.getenv('SCRAPER_API_KEY')
```

```python
import requests
import json

# Set your search query
search_query = "violencia"

# Set your initial and end dates
initial_date = "2016-01-01"
end_date = "2023-01-01"

# Set your latitude and longitude
latitude = 2.4411
longitude = -76.6061

# Create a Scraper API request URL
api_url = "http://api.scraperapi.com/structured/twitter/search"

# Make the request
try:
    response = requests.get(api_url, params={
        "api_key": scraper_api_key,
        "query": search_query,
        "ini": initial_date,
        "end": end_date,
        "latitude": latitude,
        "longitude": longitude
    })
except Exception as e:
    print(e)

# Parse the JSON response
data = json.loads(response.content)
```

```python
twitter_data = []

# Extract the tweets
tweets = data["organic_results"]

# Iterate over the tweets and print the text
for tweet in tweets:
    twitter_data.append(tweet)
```
