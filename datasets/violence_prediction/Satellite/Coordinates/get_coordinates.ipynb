### Get municipalities and coordinates

```python
import pandas as pd
```

```python
# Read the original csv file
df = pd.read_csv('../../Violence/Census_and_Domestic_Violence.csv')
df
```

```python
# Take just cases from 2016 - 2018
df_cases_2016_2023 = df.iloc[:,372:]
# Add Municipality, Municipality code
df_cases_2016_2023[['Municipality', 'Municipality code']] = df[['Municipality', 'Municipality code']]
# Use the Municipality code as index
df_cases_2016_2023.set_index(['Municipality code'], inplace=True)

df_cases_2016_2023
```

```python
# Create a new dataframe to calculate cases by municipality from 2016 to 2018
# Sum all cases and sort using the sum of cases
cases_by_municipality = (df_cases_2016_2023.iloc[:, :-1].sum(axis=1)).sort_values(ascending = False).reset_index()
cases_by_municipality.rename(columns={0:'cases_number'}, inplace=True)
cases_by_municipality
```

```python
# Add the cases to the dataset with cases by week, Municipality and department using the Municipality code
cases_by_municipality = cases_by_municipality.merge(df_cases_2016_2023, on=['Municipality code'], how='inner')
cases_by_municipality[['Municipality', 'Municipality code', 'cases_number']]
```

## Take latitude and longitude

```python
# Import google earth engine API
import ee
import geemap
```

```python
# Trigger the authentication flow.
ee.Authenticate()

# Initialize the library.
ee.Initialize()
```

# Constants

```python
# Width and Heigth [px]
LENGTH = 512

# Resolution of satellite [px/m] (10 if Sentinel)
RESOLUTION = 10
```

```python
# See a map
Map = geemap.Map(center=[42.3601, -71.0589], zoom=13)
Map.add_basemap('SATELLITE')
Map
```

```python
import numpy as np
# Get rectangular region from irregular figure (E.g. Circle or Polygon)
def get_rectangular_region(region):
  try:
    # If is a feature collection get the geometry and then the coordinates
    coordenates = np.array(region.geometry().getInfo()['coordinates'][0])
  except:
    # If it's a geomerty get the coordinates
    coordenates = np.array(region.getInfo()['coordinates'][0])

  latitude = []
  longitude = []
  # Get min and max value for latitud
  latitude = [coordenates[:,0].min(), coordenates[:,0].max()]
  # Get min and max value for longitude
  longitude = [coordenates[:,1].min(), coordenates[:,1].max()]


  # Build square:
  coordinates_square = [latitude[0], longitude[0], latitude[1], longitude[1]]

  return coordinates_square
```

```python
def distance_and_area(region, coordinates_square):

  # Width
  # Define a Point object.
  point = ee.Geometry.Point(coordinates_square[0], coordinates_square[1])
  # Define other inputs.
  inputGeom = ee.Geometry.Point(coordinates_square[2], coordinates_square[1])
  # Apply the distance method to the Point object.
  width = point.distance(**{'right': inputGeom, 'maxError': 1})
  #Print the result to the console.
  print('The width is =', width.getInfo())

  # Height
  # Define a Point object.
  point = ee.Geometry.Point(coordinates_square[0], coordinates_square[1])
  # Define other inputs.
  inputGeom = ee.Geometry.Point(coordinates_square[0], coordinates_square[3])
  # Apply the distance method to the Point object.
  height = point.distance(**{'right': inputGeom, 'maxError': 1})
  # Print the result to the console.
  print('The height is =', height.getInfo())

  # Area
  regionArea = region.geometry().area(**{'maxError': 1})
  print(f'The area is: {regionArea.getInfo()}')
```

# Cali

```python
# Create map instance
Map = geemap.Map()
Map.add_basemap('SATELLITE')

# Take the municipality using Municipality code
city_coordinates = df_w_coordinates[df_w_coordinates['Municipality code'] == 76001]
# Take the coordinates
centroid = city_coordinates[['lat', 'lng']].values.reshape(2,)
# Create a ee Point instance
centroid = ee.Geometry.Point(centroid[1], centroid[0])
Map.addLayer(centroid, {}, "Point")

# Get Circular Area arround the center of the city
area = centroid.buffer(LENGTH*RESOLUTION/2) # 750 px * 10 m/px = 7500 m
#Map.addLayer(area, {}, "buffered area")

# Get square area from circle
coordinates_square = get_rectangular_region(area)
print(f'The coordinates of Cali of {LENGTH} x {LENGTH} as rectangle are: {coordinates_square}')
region = ee.FeatureCollection(ee.Geometry.Rectangle(coordinates_square))
distance_and_area(region, coordinates_square)
Map.addLayer(region, {}, "Region_square")

Map.centerObject(region, zoom=12)
Map
```

# Square for each municipality

```python
def get_square_coordinates(coordinates):
  centroid = ee.Geometry.Point(coordinates[1], coordinates[0])
  # Get Circular Area arround the center of the city
  area = centroid.buffer(LENGTH*RESOLUTION/2) # E.G. 750 px * 10 m/px = 7500 m

  # Get square area from circle
  coordinates_square = get_rectangular_region(area)

  return coordinates_square
```

```python
coordinates_df = df_w_coordinates[['lat', 'lng', 'Municipality', 'Municipality code']]
# coordinates_df = coordinates_df.iloc[:10, :] # Get just top 10
coordinates_df
```

```python
coordinates_df['square'] = coordinates_df[['lat', 'lng']].apply(get_square_coordinates, axis=1)
coordinates_df
```

```python
dict_coordinates = coordinates_df[['Municipality', 'square']].iloc[:10, :]
dict_coordinates
```
