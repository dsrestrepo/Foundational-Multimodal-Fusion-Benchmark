```python
import os
import numpy as np
import pandas as pd

from utils import normalize_embeddings, visualize_embeddings
```

```python
def calculate_var(df, DATASET='', DIR='No_shift', save=False):
    # Select features and labels vectors
    text_columns = [column for column in df.columns if 'text' in column]
    image_columns = [column for column in df.columns if 'image' in column]

    # Normalize initial embeddings
    text_embeddings = df[text_columns].values
    image_embeddings = df[image_columns].values

    # Visualize shifted embeddings
    visualize_embeddings(text_embeddings, image_embeddings, f'Embeddings for {DATASET}', lambda_shift=0, DATASET=DIR+f'_{DATASET}', save=save, var=True)
    
```

### Medical Datasets

### BRSET

```python
PATH = 'Embeddings_vlm/brset/'
FILE = 'embeddings_clip.csv'

df = pd.read_csv(os.path.join(PATH, FILE))
df.drop(columns=['image_id', 'text'], inplace=True)

calculate_var(df, 'BRSET', save=True)
```

### HAM 10000

```python
PATH = 'Embeddings_vlm/ham10000/'
FILE = 'embeddings_clip.csv'

df = pd.read_csv(os.path.join(PATH, FILE))
df.drop(columns=['image_id', 'text'], inplace=True)

calculate_var(df, 'HAM 10000', save=True)
```

### SatelliteBench

```python
PATH = 'Embeddings_vlm/satellitedata/'
FILE = 'embeddings_clip.csv'

df = pd.read_csv(os.path.join(PATH, FILE))
df.drop(columns=['image_id', 'text'], inplace=True)
df.Labels = df.Labels.astype(str)

calculate_var(df, 'SatelliteBench', save=True)
```
